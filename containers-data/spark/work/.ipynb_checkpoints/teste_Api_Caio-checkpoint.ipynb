```python
from pyspark.sql import SparkSession
from delta.tables import DeltaTable
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

# Cria uma sessão Spark
spark = SparkSession.builder \
    .appName("DeltaTableCreateIfNotExistsExample") \
    .config("spark.jars.packages", "io.delta:delta-core_2.12:2.0.0") \
    .getOrCreate()

# Define o caminho onde a Delta Table será armazenada
delta_table_path = "/path/to/delta_table"

# Define o esquema para a Delta Table
schema = StructType([
    StructField("name", StringType(), nullable=False),
    StructField("age", IntegerType(), nullable=False)
])

# Cria um DataFrame com dados exemplo
data = [("John", 28), ("Jane", 33), ("Joe", 40)]
df = spark.createDataFrame(data, schema)

# Define as propriedades da Delta Table
delta_table = DeltaTable.createIfNotExists(spark) \
    .tableName("example_table") \
    .addColumn("name", "STRING") \
    .addColumn("age", "INT") \
    .location(delta_table_path) \
    .execute()

print("Delta Table criada se não existir!")

# Para salvar o DataFrame na Delta Table
df.write.format("delta").mode("append").save(delta_table_path)

# Para ler a Delta Table
df_loaded = spark.read.format("delta").load(delta_table_path)
df_loaded.show()
```
